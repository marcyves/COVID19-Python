# COVI19 Analyse

This script uses data generated by http://blog.lazd.net/coronadatascraper/

```python
import pandas as pd

df = pd.read_csv('http://blog.lazd.net/coronadatascraper/data.csv')

# Compute death rate
df['rate'] = df['cases']/df['deaths']
df_sorted = df.sort_values('cases', ascending=False)
df_sorted
```

```python
df_france = df_sorted[df_sorted['country'] == 'France']
df_france
```

```python
# Extract only Country data
# only_country = df[df[['county', 'state']].isnull().all(1)]
only_country = df[df['county'].isnull()]

# sort by number of cases
only_country_sorted = only_country.sort_values('cases', ascending=False)
only_country_sorted
```

```python
# Compute death rate
only_country_sorted['rate'] = only_country_sorted['cases']/only_country_sorted['deaths']
only_country_sorted
```
